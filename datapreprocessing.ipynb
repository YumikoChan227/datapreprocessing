```python
import pandas as pd
```

```python
dataset = pd.read_csv('loan_small.csv')
```

```python
df = cleandata.copy()
```

```python
col = ['Gender', 'Area', 'Loan_Status']
df[col] = df[col].fillna(df.mode().iloc[0])
```

```python
df.isnull().sum(axis=0)
```

```python
cols = ['ApplicantIncome', 'CoapplicantIncome','LoanAmount']
df[cols] = df[cols].fillna(df.mean())
```

```python
df.isnull().sum(axis=0)
```

```python
df = df.drop(['Loan_ID'], axis=1)
```

```python
df
```

```python
df = pd.get_dummies(df, drop_first=True)
```
